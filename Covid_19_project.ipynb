<p><b> 4 - Filter the dataframe with only this columns 
 
['continent','location','date','total_cases','total_deaths','gdp_per_c

```python
columns_to_keep = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']
df = df[columns_to_keep]
df
```

<p> <b>5. Data Cleaning<br>
a. Remove all duplicates observations
b. Find missing values in all columns
c. Remove all observations where continent column value is
missing
Tip : using subset parameter in dropna
d. Fill all missing values </b></p>with 0

```python
df = df.drop_duplicates()   # (a)
df
```

```python
missing_values = df.isnull().sum()  # (b)
missing_values
```

```python
df = df.dropna(subset=['continent'])   # (c)
df
```

```python
df = df.fillna(0)  # (d)
df
```

<p><b>6. Date time format :<br>
a. Convert date column in datetime format using
pandas.to_datetime

b. Create new column month after extracting month data from
date
column.</b></p>

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')   # (a)
df
```

```python
df['month'] = df['date'].dt.month  # (b)
df['month']
```

<p><b> 7. Data Aggregation:<br>
a. Find max value in all columns using groupby function on
'continent'
column
Tip: use reset_index() after applying groupby
b. Store the result in a new dataframe named 'df_groupby'.
(Use df_groupby dataframe for all further analysis)</b></p>

```python
df_groupby = df.groupby('continent').max().reset_index() # (a) & (b)
df_groupby
```

<p><b> 
    8. Feature Engineering :<br>
a. Create a new feature 'total_deaths_to_total_cases' by ratio of
'total_deaths' column to 'total_cases'
</b></p>

```python
df_groupby['total_deaths_to_total_cases'] = (df_groupby['total_deaths'] / df_groupby['total_cases'])
df_groupby['total_deaths_to_total_cases'] = (df_groupby['total_deaths_to_total_cases'].fillna(0).replace([float('inf'), -float('inf')], 0)) * 100
df_groupby
```

<p><b>
    9. Data Visualization :<br>
a. Perform Univariate analysis on 'gdp_per_capita' column by
plotting
histogram using seaborn dist plot.
b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
c. Plot Pairplot on df_groupby dataset.
d. Plot a bar plot of 'continent' column with 'total_cases' .
Tip : using kind='bar' in seaborn catplot
</b></p>

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['gdp_per_capita'], kde=True, bins=30, color='blue')    # (a)
plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.jointplot(data=df_groupby, x='gdp_per_capita', y='total_cases', kind = 'scatter')  # (b)
plt.title('Scatter Plot of Total Cases vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
plt.show()
```

```python
sns.pairplot(df_groupby)   # (c)
plt.suptitle('Pairplot of Aggregated Data (df_groupby)', y=1.02)
plt.show()
```

```python
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar', aspect=2, palette='viridis')  # (d)
plt.title('Total Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```
